```python
import os
import json
from collections import defaultdict
import copy
```

```python
f = open('./vansh_phrases.json')
phrases = []

temp = json.load(f)
for t in temp:
    transcript = t['transcript']
    phrases.append(transcript)

print(len(phrases))

vocab = defaultdict(int)
for phrase in phrases:
    for word in phrase.split():
        vocab[word] += 1

vocab_keys = sorted(vocab.keys())
print(len(vocab_keys))
print(vocab_keys)

# with open('vocab1.txt', 'w') as file:
#     file.write('\n'.join(vocab_keys))
```

```python
lines = open('./claude-sentences.txt', 'r').readlines()
word_freq = defaultdict(int)

for line in lines:
    words = line.lower().split()
    for word in words:
        word_freq[word.replace(',', '')] += 1

vocab = sorted(word_freq.keys())
print(len(vocab))
print(vocab)

with open('./claude-vocab.txt', 'w') as file:
    file.write('\n'.join(vocab))
```

```python
# creating a word cloud
import matplotlib.pyplot as plt
from wordcloud import WordCloud

claude_word_freq = defaultdict(int)
with open('./claude-sentences.txt', 'r') as file:
    for line in file.readlines():
        for word in line.lower().split():
            claude_word_freq[word.replace(',', '')] += 1


# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(claude_word_freq)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
